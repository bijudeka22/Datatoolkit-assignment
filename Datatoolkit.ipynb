#2.How does broadcasting work in NumPy

NumPy broadcasting is a powerful feature that allows you to perform element-wise operations on arrays with different shapes, as long as their dimensions are compatible. It avoids the need for explicit loops, making your code more concise and efficient.
Here's how it works:
 * Dimension Compatibility: NumPy compares the shapes of the two arrays element-wise, starting from the trailing dimensions and working backward. For two dimensions to be compatible, they must either be equal or one of them must be 1.
 * Stretching Dimensions: If one of the arrays has a dimension of 1, it is "stretched" or "broadcasted" to match the shape of the other array along that dimension.
 * Element-wise Operation: Once the shapes are compatible, NumPy performs the element-wise operation as if the arrays had the same shape.

#17.Explain the difference between apply() and map() in Pandas?
- In Pandas, both apply() and map() are used to apply functions to data, but they operate at different levels:
 * map(): Works element-wise on a Series.  It applies a function to each individual value in the Series.
 * apply(): Can work on a Series or a DataFrame.  On a Series, it can behave like map(), but it can also perform more complex operations. On a DataFrame, it can apply a function to each row or each column.
In short: map() is for simple, element-wise transformations on a Series. apply() is more versatile and can be used for more complex operations on both Series and DataFrames.

#21.Why is NumPy’s array slicing faster than Python’s list slicing?
- NumPy's array slicing is generally faster than Python's list slicing due to a few key reasons:
 * Contiguous Memory: NumPy arrays are stored contiguously in memory, meaning elements are placed next to each other. This allows for direct access to slices without needing to follow pointers, making retrieval much faster. Python lists, on the other hand, store pointers to objects scattered in memory, requiring more overhead to gather a slice.
 * Vectorized Operations: NumPy leverages vectorized operations, which means operations are performed on entire arrays at once rather than element by element. This is implemented in highly optimized C code, leading to significant performance gains. Slicing in NumPy often utilizes these optimized operations.
 * Views vs. Copies: NumPy slicing often creates "views" of the original array rather than making a full copy of the data. This means that changes to a slice can affect the original array, but it also avoids the overhead of copying large amounts of data. Python list slicing always creates a new copy.
In essence, NumPy's design for numerical computation, with its contiguous memory layout and vectorized operations, makes array slicing significantly more efficient than Python's list slicing, especially for large datasets.
